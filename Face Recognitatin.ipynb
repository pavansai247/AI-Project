# Encodes of the Face images

```python
import face_recognition
import cv2
import os
import pickle
folder_path = r"D:\Pavan sai\Datascience\OpenCV\Face Detection\Face images"
pathlist = os.listdir(folder_path)
imglist = []
names = []
for path in pathlist:
    imglist.append(cv2.imread(os.path.join(folder_path,path)))
    names.append(os.path.splitext(path)[0])
```

```python
def findencodings(imageslist):
    face_encodings =[]
    for img in imageslist:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        encode = face_recognition.face_encodings(img)[0]
        face_encodings.append(encode)
    return face_encodings
```

```python
encodings =findencodings(imglist)
```

```python
encodewithnames = [encodings,names]
encodewithnames
```

```python
file = open("encodingfile.p", 'wb')
pickle.dump(encodewithnames, file)
file.close()
```
